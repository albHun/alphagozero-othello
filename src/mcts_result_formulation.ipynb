```python
from pprint import pprint
import json
import numpy as np
```

```python
data = json.load(open('0', 'r'))
```

```python
pprint(data)
```

```python
winner = data[1]
if winner == 2:
    winner = -1
print(winner)
```

```python
mcts = data[0]
pprint(mcts)
```

```python
X, y = [], []
```

```python
for i in range(len(mcts)):
    state = json.loads(mcts[:][i][0])
    print("Current state:")
    state = np.array(state)
    np.place(state, state==2, -1)
    np.reshape(state, (state.shape[0], state.shape[1], 1))
    X.append(state)
    pprint(state)
    
    ms = mcts[:][i][2]
    grid_ps = np.zeros(np.array(state).shape)
    ps = mcts[:][i][3]
    if ms != str(None):
        for i in range(len(ms)):
            position = ms[i]
            grid_ps[position[0], position[1]] = ps[i]
        y.append(np.concatenate((grid_ps.flatten(), np.array([winner])), axis = 0))
        print("Probabilities of moves:")
        pprint(grid_ps)
        player = mcts[:][i][1]
    else:
        y.append(np.concatenate((grid_ps.flatten(), np.array([winner])), axis = 0))
print(winner)
```

```python
X = np.array(X)
y = np.array(y)
```
